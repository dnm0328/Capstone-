```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
def single_season_comparison(string):
    '''
    DEFINITION: a function that reads in a
    csv file and converts it to a histogram 
    after first running a welch's t-test between
    the two samples. 
    
    --------------------------------
    
    PARAMETERS: string = path to csv file
    
    --------------------------------
    
    RETURNS: Histogram comparing the normal
    distributions of playoff/ non-playoff 
    teams' mean yds/ points for a single seas
    against the over all defensive means. 
    '''
    string_lst = ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
    for i in string_lst:
        if i in string:
            year = i
    print(year)
    df = pd.read_csv(string)
    filtered = df[['year', 'Team', 'Pts/G', 'Yds/G', 'Playoff']]
    total_yrds = df['Yds/G'].sum()
    count = len(filtered['Team'])
    total_avg_yrds = total_yrds/ count
    playoff_teams = filtered.loc[filtered.Playoff==True]
    no_playoff_teams = filtered.loc[filtered.Playoff==False]
    mean1 = playoff_teams['Yds/G'].sum()/ playoff_teams['Yds/G'].count()
    mean2 = no_playoff_teams['Yds/G'].sum()/ no_playoff_teams['Yds/G'].count()
    fig, ax = plt.subplots(1, figsize=(12, 5))
    ax.vlines(x=total_avg_yrds, ymin=0, ymax=.022, linewidth=4, linestyles='solid', label='Cumulative mean', color='yellow')
    ax.vlines(x=mean1, ymin=0, ymax=.022, linewidth=3, linestyles='solid', label='Playoff mean', color='b')
    ax.vlines(x=mean2, ymin=0, ymax=.022, linewidth=3, linestyles='solid', label='Non-Playoff mean', color='r')
    ax.hist(playoff_teams['Yds/G'],alpha=.7, bins=6, density=True, label="Teams that Made Playoffs")
    ax.hist(no_playoff_teams['Yds/G'], alpha=.4,bins=6, density=True, label="Teams that Didn't Make Playoffs")
    ax.set_xlabel('Allowed Yards/ Game')
    ax.legend()
    ax.set_title("Distribution of 'AVG_YDS Against' for the Season of {}".format(year))
```

```python
single_season_comparison('data/2009_Update.csv')
```
