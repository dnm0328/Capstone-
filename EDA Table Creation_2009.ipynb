```python
import pandas as pd
```

***Turn NFL full-season defensive data into cleaned tables representing teams that made/ didn't make playoffs (two tables)***

```python
df_2009 = pd.read_csv('data/2009.csv')
```

```python
df_2009.head()
```

```python
filtered_2009 = df_2009[['Team', 'Gms', 'Pts/G', 'Yds/G', 'year']]
```

```python
filtered_2009
```

```python
Playoff = ['True', 'True', 'True', 'True', 'False', 'True', 'False', 'False', 'True', 'False', 'True', 'True', 'False', 'False', 'False', 'True', 'False', 'True', 'False', 'True', 'False', 'False', 'False', 'False', 'True', 'False', 'False', 'False', 'False', 'False', 'False', 'False']
```

```python
filtered_2009['Playoff'] = Playoff
```

```python
filtered_2009.head(10)
```

```python
Playoff_tb = filtered_2009.loc[lambda filtered_2009: filtered_2009['Playoff'] == 'True']
Playoff_tb.head(5)
```

```python
no_Playoff_tb = filtered_2009.loc[lambda filtered_2009: filtered_2009['Playoff'] == 'False']
no_Playoff_tb.head(5)
```

***Get averaged data for both playoff/ non-playoff tables for future comparisons (2009)***

```python
pts_avg = Playoff_tb['Pts/G'].mean()
pts_avg = round(pts_avg, 2)
yds_avg = Playoff_tb['Yds/G'].mean()
yds_avg = round(yds_avg, 2)
print('The teams that made the playoffs allowed {:2.2f} points/ game'.format(pts_avg))
print('The teams that made the playoffs allowed {:2.2f} yards/ game'.format(yds_avg))
```

```python
pts_avg2 = no_Playoff_tb['Pts/G'].mean()
pts_avg2 = round(pts_avg2, 2)
yds_avg2 = no_Playoff_tb['Yds/G'].mean()
yds_avg2 = round(yds_avg2, 2)
print('The teams that did NOT make the playoffs allowed {:2.2f} points/ game'.format(pts_avg2))
print('The teams that did NOT make the playoffs allowed {:2.2f} yards/ game'.format(yds_avg2))
```

```python
data = {'avg_stats': ['allowed_pts/G', 'allowed_yds/G'], 
        'Playoff_teams': [pts_avg, yds_avg],
        'non_Playoff_teams': [pts_avg2, yds_avg2]}
diff_2009 = pd.DataFrame(data=data)
diff_2009.set_index('avg_stats', inplace=True)
diff_2009
```

```python
minus = diff_2009['non_Playoff_teams'] - diff_2009['Playoff_teams']
data_sub = round(minus, 2)
diff_2009['Difference'] = data_sub
```
